# nifH MAGs analysis

This code is used to associate putative gene sequence hits from the HMMER searches of the metagenomes with the high quality MAGs they are found in.

```python
import pandas as pd
```

```python
import os
import glob
import matplotlib.pyplot as plt
```

```python
#define key paths. Gene name, results path, and output folder can be changed as necessary to accommodate different searches.
gene= "nifH"
mag_results_path=os.path.join("/vortexfs1/omics/alexander/lblum/tara_gene_finder/results/mag_results",gene+".maghits.contigs.csv")
output_folder= "/vortexfs1/omics/alexander/lblum/tara_gene_finder/jupyter_notebooks/outputs/nifH/"
```

```python
#Quality cutoffs were set to 80% complete and less than 5% contamination as predicted by CheckM
high_complete= 80
high_contam = 5
```

```python
#table with bins and associated contigs for each MAG
mag_results=pd.read_csv(mag_results_path , index_col=0)
```

```python
#MAGs CheckM report
mags_checkm= pd.read_csv("/vortexfs1/omics/alexander/lblum/tara_gene_finder/data/mags_data/all_checkm_plus.txt",sep='\t', index_col=0)
```

```python
mags_checkm[(mags_checkm.contamination <= 5)& (mags_checkm.completeness >= 80)]
```

```python
#MAGs quast report
mags_quast = pd.read_csv('/vortexfs1/omics/alexander/lblum/tara_gene_finder/data/mags_data/all_quast_report.txt', sep='\t', index_col=0)
```

```python
#gives us an idea of taxonomy; diamond search "BLAST" against genomes in UniProt
mags_diamond_bin= pd.read_csv('/vortexfs1/omics/alexander/lblum/tara_gene_finder/data/mags_data/all_diamond_bin_report.tsv', sep='\t', index_col=0)
```

```python
mags_checkm_subset= mags_checkm.loc[mag_results.index].dropna(how= "all")
```

```python
num_columns=['completeness', 'contamination', 'strain_het']
```

```python
mags_checkm_subset[num_columns] = mags_checkm_subset[num_columns].apply(pd.to_numeric, errors='coerce')
```

```python
#visually assess the completeness and contamination of MAGs
fig, axs = plt.subplots(1)
mags_checkm_subset.plot(kind="scatter", x="completeness", y='contamination', ax=axs)
axs.set_ylim(0,50)
```

```python
#MAGs which meet completeness and contamination standards are filtered out
high_quality_mags = mags_checkm_subset[(mags_checkm_subset.contamination <= high_contam)& (mags_checkm_subset.completeness >= high_complete)]
```

```python
#this list of high quality MAGs and metadata is sent to csv
high_quality_mags.to_csv(os.path.join(output_folder,gene+"_high_quality_mags_info.csv"))
```

```python
#Here we get our simple list of hig_quality_mags with the bin_ID for our downstream mag analysis
high_quality_mags['gene']=gene
high_quality_mags_list=high_quality_mags['gene']
high_quality_mags_list.to_csv(os.path.join(output_folder,gene+"_high_quality_mags.csv"))
```
